```python
import pandas as pd
from sklearn.model_selection  import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
data = pd.read_csv("../data/guardian/knn_model/final_files/1476371065-100-train.csv")

x = data.iloc[:,1:]
y = data.iloc[:,0:1].values.ravel()

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
n_neighbors = 11
neigh = KNeighborsClassifier(n_neighbors, weights='distance')
neigh.fit(X_train, y_train)
y_pred = neigh.predict(x)
print(confusion_matrix(y,y_pred))
print(classification_report(y,y_pred))
```

```python
from collections import Counter
#0.1Attack
# data = pd.read_csv("../data/guardian/knn_model/knn_training_8331285754-300-1.csv")
# data = pd.read_csv("../data/guardian/knn_model/test_1conv_8331285754-300-1.csv")
#1Attack
# data = pd.read_csv("../data/guardian/knn_model/knn_training_9387700481-300-1.csv")
# data = pd.read_csv("../data/guardian/knn_model/test_1conv_9387700481-300-1.csv")
#5Attack
# data = pd.read_csv("../data/guardian/knn_model/knn_training_7117532915-300-4.csv")
data = pd.read_csv("../data/guardian/knn_model/test_1conv_7117532915-300-4.csv")
#10Attack
# data = pd.read_csv("../data/guardian/knn_model/test_1conv_8864816883-300-2.csv")

# Make a copy of the DataFrame to keep the original data unchanged
df_copy = data.copy()

# Step 1: Calculate the median of each row and save it into a new column
df_copy['median'] = df_copy.median(axis=1)

# Step 2: Save the highest value of each row in a new column
df_copy['max_value'] = df_copy.max(axis=1)

# Step 3: Compute the distance between the highest value and median
df_copy['dist_median_max'] = df_copy['max_value'] - df_copy['median']

# Step 4: Write the function for conditional labeling
def conditional_label(row):
    if (row['median'] > 0.2 and row['dist_median_max'] > 0.2) or row['max_value'] > 0.5:
        return True
    else:
        return False

# Apply the function to each row
df_copy['label'] = df_copy.apply(conditional_label, axis=1)

# A helper function to process each batch
def process_batch(batch):
    # Count the number of TRUE values in the batch
    true_count = batch.sum()
    # Make a decision based on the count of TRUE values
    return "ATTACK" if true_count >  8 else "SAFE"

# Step 2 & 3: Process batches and make decisions
# Apply the function to each batch of 10 rows in the 'label' column
# Note: Adjust the batch size if necessary by changing '10' to another value
results = [process_batch(df_copy['label'][i:i+10]) for i in range(0, len(df_copy['label']), 10)]

# Step 4: Store results
# Here we just print the results, but you could also add them to your DataFrame or save them to a file
# print(results[0:88])
# print(results[0:22])
print(f'Among real attackes attack is: {results[0:22].count("ATTACK")}')
print(f'False attack is: {results[22:].count("ATTACK")}')
print(f'Total Attack is: {results.count("ATTACK")}')
print(f'Safe is: {results.count("SAFE")}')
print(f"Total number of accounts is {len(results)}, which includes {Counter(results)}.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the user's table
models = ['Scenario 1', 'Scenario 2']
metrics = ['Precision', 'Recall', 'Val_precision', 'Val_recall', 'F1', 'Computation Time']
scenario_1_values = [0.99, 0.99, 0.80, 0.82, 0.81, 15605]  # Including computation time for Scenario 1
scenario_2_values = [0.99, 0.99, 0.73, 0.84, 0.81, 22091]  # Including computation time for Scenario 2

bar_width = 0.35  # width of the bars
index = np.arange(len(metrics))  # the label locations

fig, ax = plt.subplots(figsize=(14, 6))  # Adjusted figure size for the additional metric

# Plotting the metrics except for computation time
bar1 = ax.bar(index[:-1] - bar_width/2, scenario_1_values[:-1], bar_width, label='Scenario 1', color='skyblue')
bar2 = ax.bar(index[:-1] + bar_width/2, scenario_2_values[:-1], bar_width, label='Scenario 2', color='green')

# Plotting computation time on a secondary axis
ax2 = ax.twinx()  # Create a second y-axis
bar3 = ax2.bar(index[-1] - bar_width/2, scenario_1_values[-1], bar_width, color='blue')
bar4 = ax2.bar(index[-1] + bar_width/2, scenario_2_values[-1], bar_width, color='darkgreen')

# Adjust font sizes and weights
ax.set_xlabel('', fontsize=20, fontweight='bold')  # Double size and bold
ax.set_ylabel('', fontsize=20, fontweight='bold')
# ax2.set_ylabel('', fontsize=20, fontweight='bold')  # Label for the secondary y-axis
ax.set_title('', fontsize=20, fontweight='bold')
ax.set_xticks(index)
ax.set_xticklabels(metrics, fontsize=20, fontweight='bold')  # Adjusted font size for readability
# ax.legend(loc='upper center', fontsize=20)

# Adding annotations to bars with adjusted size and weight
def add_labels(bars, ax_used):
    for bar in bars:
        height = bar.get_height()
        ax_used.annotate('{:.2f}'.format(height),
                         xy=(bar.get_x() + bar.get_width() / 2, height),
                         xytext=(0, 3),  # 3 points vertical offset
                         textcoords="offset points",
                         ha='center', va='bottom',
                         fontsize=20,  # Adjusted font size for readability
                         fontweight='bold')

add_labels(bar1, ax)
add_labels(bar2, ax)
# For computation time, you might want to format the annotation differently
add_labels(bar3, ax2)
add_labels(bar4, ax2)

plt.savefig('model_performance.png', format='png', dpi=700)  # Raster format, high resolution

plt.tight_layout()
plt.show()


```
